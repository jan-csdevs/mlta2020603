## Data Set

The data set contains input parameters $r_y$, $r_z$, $cr_x$ to a quantum ansatz and resulting state complex state vectors, comprising the compelx components $S_{00}$, $S_{01}$, $S_{10}$, $S_{11}$.

Here, we will only be interested in the real parts of the state vector components.

```python
import numpy as np
import pandas as pd
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from random import random,seed
from math import pi
```

```python
seed(42)

output = []
v_in = []

for _ in range(1000):
    v = [random()*2*pi for _ in range(3)]
    v_in.append(v)
    circuit = QuantumCircuit(2)
    circuit.ry(v[0],0)
    circuit.rz(v[1],1)
    circuit.cx(0,1)
    circuit.crx(v[2],0,1)
    state_vector = Statevector.from_instruction(circuit)
    output.append(state_vector.data)
```

```python
v_in = np.array(v_in)
output = np.array(output)
data = pd.DataFrame({"ry": v_in[:,0],
                     "rz": v_in[:,1],
                     "crx": v_in[:,2],
                     "S00": output[:,0],
                     "S01": output[:,1],
                     "S10": output[:,2],
                     "S11": output[:,3],
})
```
